```python
import matplotlib.pyplot as plt
from ultralytics import YOLO
from glob import glob

path_images = glob(r'Preparando os dados\mosaico_Centro\*.jpg')

for path_image in path_images:

    model = YOLO(r'Treinamento da IA\runs\detect\runs\exp1\weights\best.pt')
    results = model.predict(
        source=path_image,
        save=True,
        conf=0.25
    )


    fig, (left, right) = plt.subplots(1, 2, figsize=(12, 6))
        
    left.imshow(plt.imread(path_image))
    left.axis('off') # Esconde os eixos numéricos
    left.set_title("Imagem Original")


    plt.imshow(results[0].plot()[:, :, ::-1])
    right.axis('off') # Esconde os eixos numéricos
    right.set_title("Detecção de Heliponto")
    plt.show()
```
